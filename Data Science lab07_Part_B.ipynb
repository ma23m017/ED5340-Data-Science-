## Question 5

WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.

## Solution

```python
# Importing the essential libraries

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Generating 't' which will be used as parameter
t = np.linspace(0, 10*np.pi, 1000)


# Calculating x, y, and z coordinates for the helical wave
x = np.sin(t)
y = np.cos(t)
z = t
```

```python
# Scatter plot method for plotting the helical wave
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='o', label='Scatter method')
ax.set_xlabel('X') 
ax.set_ylabel('Y')
ax.set_zlabel('Z') 
ax.legend() 

# Displaying the scatter plot
plt.show()
```

```python
# Line plot method for plotting the helical wave
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, c='b', label='Normal line method')
ax.set_xlabel('X') 
ax.set_ylabel('Y') 
ax.set_zlabel('Z') 
ax.legend() 

# Displaying the line plot
plt.show()
```

## Question 8

Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy.

## Solution

```python
import numpy as np


N, M = map(int, input("Enter number of rows and columns (N M): ").split())
matrix = []
for _ in range(N):
    row = list(map(int, input().split()))
    matrix.append(row)

# Convert the input list into a numpy array
arr = np.array(matrix)

# Print the original array
print("Original Array:")
print(arr)

# Transpose the array
transpose_arr = np.transpose(arr)

# Print the transposed array
print("\nTransposed Array:")
print(transpose_arr)

# Flatten the array
flatten_arr = arr.flatten()

# Print the flattened array
print("\nFlattened Array:")
print(flatten_arr)
```

## Question 9

WAP to capitalize a column of names in a Pandas Dataframe.

Eg : Input : {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

Output : {'Name': ['John', 'Boday', 'Ana', 'Peter', 'Nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

## Solution

```python
import pandas as pd

# Input data
data = {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'],
        'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'],
        'Age': [27, 23, 21, 23, 24]}

# Create DataFrame
df = pd.DataFrame(data)

# Capitalize the 'Name' column
df['Name'] = df['Name'].str.capitalize()

# Output the DataFrame
print(df)
```

## Question 10

 Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of “Optimization for Engineering Design: Algorithms and Examples” by KALYANMOY DEB, 2nd edition


f(x) = 3x**2 + 2x

## Solution

```python
import numpy as np

# Define the function
def f(x):
    return 3*x**2 + 2*x

# Define the point at which to compute the derivatives
x = 2

# Define the step size for the central difference method
h = 0.01

# Compute the first-order derivative using central difference method
df_dx_1 = (f(x + h) - f(x - h)) / (2 * h)

# Compute the second-order derivative using central difference method
df_dx_2 = (f(x + h) - 2 * f(x) + f(x - h)) / (h ** 2)

# Print the results
print("First-order derivative at x =", x, ":", df_dx_1)
print("Second-order derivative at x =", x, ":", df_dx_2)
```
