## Question 1

Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

## Solution

```python
#importing essential libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Taking the value of coefficients from the user

A = float(input("Enter coefficient A: "))
B = float(input("Enter coefficient B: "))
C = float(input("Enter coefficient C: "))
D = float(input("Enter coefficient D: "))
E = float(input("Enter coefficient E: "))

coefficients = [A, B, C, D, E]  #storing the value of coefficients in a list

print('Entered coefficients are:\n', coefficients)
```

```python
#defining a function to calculate the polunomial values

def calculate_polynomial(coefficients, x_values):
    A, B, C, D, E = coefficients
    polynomial_values = A * x_values**4 + B * x_values**3 + C * x_values**2 + D * x_values + E
    return polynomial_values
```

```python
x_values = np.arange(-100, 101) # Generate x values from -100 to 100
y_values = calculate_polynomial(coefficients, x_values) # Values of y for the corresponding values of x

#storing the entered values of x and y in numpy array
x_values = np.array(x_values)
y_values = np.array(y_values)
```

```python
#ploting the graph using generated numpy array

plt.plot(x_values, y_values)
plt.title('Graph of the Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Question 2

Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.

## Solution

```python
# Importing essential libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Given dictionary containing sales data

sales_data = {

'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

#print(sales_data)
```

```python
# Converting dictionary to pandas

df = pd.DataFrame(sales_data)
print(df)
```

```python
# Calculating total sales for each product

total_sales = df.groupby('Product')['Sales'].sum().reset_index()
print(total_sales)
```

```python
# Plotting the bar chart

plt.bar(total_sales['Product'], total_sales['Sales'])
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales for Each Product')
plt.show()
```

## Question 3

 Create visualizations for the following mathematical functions using Matplotlib:

Plot the following single-variable functions over the range 

[10,10], and include a title and labels for the axes:

(1) y = cos(x)

(2) y = e^x

(3) y = log(x), where x>0


Generate surface plots for these multi-variable functions over the range 

x=[10,10] and y=[10,10] , ensuring to add a title and labels for all axes:

(1) z = cos(sqrt(x^2+y^2)

(2) z = e^(-(x^2+y^2))

(3) z =  log(x^2+y^2) where x^2+y^2>0

## Solution

```python
# Importing important libraries

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Single variable functions

```python
#taking values from defined range

x = np.linspace(-10, 10, 400)
```

### (1) y = cos(x)

```python
y = np.cos(x)

plt.plot(x, y)
plt.title('y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### (2) y = e^x

```python
y = np.exp(x)

plt.plot(x, y)
plt.title('y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### (3) y = log(x)

```python
y = np.log(x[x>0])

plt.plot(x[x>0], y)
plt.title('y = log(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### Multivariable functions

```python
# Taking the values of x and y for plotting the multivariable functions

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
```

### (1) z = cos(sqrt(x^2+y^2)

```python
z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, z)
ax.set_title('z = cos(sqrt(x^2 + y^2))')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### (2)  z = e^(-(x^2+y^2))

```python
z = np.exp(-(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, z)
ax.set_title('z = e^(-(x^2+y^2))')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

### (3) z = log(x^2+y^2) where x^2+y^2>0

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x[x!=0], y[y!=0])

z = np.log(X**2 + Y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, z)
ax.set_title('z = log(x^2+y^2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

## Question 4

 For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).

## Solution

```python
def J(w):
    return w**2 + 54 / w

def minimum_interval_search(J, a, b, n):
    dw = (b - a) / n
    
    w1 = a
    w2 = w1 + dw
    w3 = w2 + dw
    
    while w3 <= b:
        if J(w1) >= J(w2) <= J(w3):
            return w1, w3
        else:
            w1 = w2
            w2 = w3
            w3 = w2 + dw
    
    # Check if minimum could be at the boundary
    if J(w1) < J(w2):
        return a, w2
    elif J(w2) < J(w3):
        return w2, b
    else:
        print("No minimum exists between (a, b)")
        return None

# Initial interval
a = 1
b = 10
n = 100

# Find the minimum interval
min_interval = minimum_interval_search(J, a, b, n)

if min_interval:
    print("Minimum interval:", min_interval)
```
